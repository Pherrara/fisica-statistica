```python
import numpy as np
import matplotlib.pyplot as plt



n = 100
epsc = 2 / (n + 1)

eps = 20 * epsc

# setting initial conditions
ka = np.zeros(n)

pt = 0.5
for i in range(n):
    r1 = np.random.rand()
    if r1 < pt:
        ka[i] = 0
    else:
        ka[i] = 1


Np = 50000

# for fast calculation max 1000000
ls = np.arange(n)

s1 = np.zeros(Np)

for it in range(Np-1):
    kax = np.random.choice(ls)
    kay = np.random.choice(ls)
    r1 = np.random.rand()
    if r1 < eps:
        r2 = np.random.rand()
        if r2 < pt:
            ka[kay] = 0
        else:
            ka[kay] = 1
    else:
        ka[kay] = ka[kax]
    s1[it+1] = np.sum(ka)

tsteps = np.arange(1, Np+1)

plt.figure(1)
plt.plot(tsteps, s1)
plt.grid()
plt.show()
```

```python
# se la eps è sotto la critica, si vedrà molto herding ovvero il sistema andrà o verso 0 o verso 100
# se la eps è sopra la critica, ci sarà molta aleatorietà ovvero il sistema ballerà intorno al 50%
# se la eps è esattamente a quella critica, ci sarà un comportamento fuoriscala, si va da 0 a 100 e viceversa più volte durante la dinamica perché non c'è un valore di k preferito rispetto ad altri
```
