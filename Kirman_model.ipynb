```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
import scienceplots # se non c'è, in terminale scrivere: pip install SciencePlots
plt.style.use(['science','notebook','grid'])
```

```python
# In questo codice applichiamo banalmente la relazione di ricorrenza del modello di Kirman in 3 casi diversi a 3 probabilità diverse
# Otterremo i 3 andamenti attesi dal modello poi in simulazioni reali
```

```python
N = 100 # abbiamo N agenti
mukl = np.zeros(N+1)   #questo è il vettore di p(k) che lavorano con la epsilon SOTTO quella critica, epsl
mukc = np.zeros(N+1)   #questo è il vettore di p(k) che lavorano con la epsilon ESATTAMENTE UGUALE quella critica, epsc
mukh = np.zeros(N+1)   #questo è il vettore di p(k) che lavorano con la epsilon SOPRA quella critica, epsh

epsc = 2/(N+1) #epsilon critica

epsl = 0.25 * epsc   #questa è la epsilon sotto
epsh = 5 * epsc #questa è la epsilon sopra

mukl[0] = 0.02
mukc[0] = 0.02
mukh[0] = 0.02
```

```python
# diamo dei valori casuali iniziali
rd = random.random()
mukl[0], mukc[0], mukh[0] = rd, rd, rd
```

```python
for i in range(N):
    mukl[i+1] = (((1-(i)/N)*(epsl/2+(1-epsl)*(i)/(N-1)))/(((i+1)/N)*(epsl/2+(1-epsl)*(1-(i)/(N-1)))))*mukl[i]
    mukc[i+1] = (((1-(i)/N)*(epsc/2+(1-epsc)*(i)/(N-1)))/(((i+1)/N)*(epsc/2+(1-epsc)*(1-(i)/(N-1)))))*mukc[i]
    mukh[i+1] = (((1-(i)/N)*(epsh/2+(1-epsh)*(i)/(N-1)))/(((i+1)/N)*(epsh/2+(1-epsh)*(1-(i)/(N-1)))))*mukh[i]
    
kx = np.arange(N+1)

mukln = mukl/np.sum(mukl)
mukcn = mukc/np.sum(mukc)
mukhn = mukh/np.sum(mukh)
```

```python
plt.plot(kx, mukln, label='Below critical probability')
plt.plot(kx, mukcn, label='Critical probability')
plt.plot(kx, mukhn, label='Above critical probability')
plt.legend()
plt.grid()
plt.show()
```

```python
#Si vede che aumentando la epsilon superiore, la binomiale si stringe
```

```python
#Allo stesso modo, diminuendo ulteriormente la epsilon inferiore, la curva diventa più sharp ed è più ripida agli estremi
```
