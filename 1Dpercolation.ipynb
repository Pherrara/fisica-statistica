```python
import igraph as ig
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
N = 300
p = 0.2

g = ig.Graph.Lattice([N], circular=True) #circular=True fa una closed chain, circular=False fa una open chain
```

```python
#ig.plot(g, layout="graphopt", vertex_size=5, bbox=(800, 450)) # questo lo plotta come un network qualsiasi, non  ottimale in questo caso

ig.plot(g, layout="circle", vertex_size=5, bbox=(800, 450)) #plottiamo il network
```

```python
# Generate random numbers and delete edges
rand = np.random.rand(g.ecount()) #generiamo un'array con tanti numeri random quanti sono i link possibili

g.delete_edges([e.index for e in g.es if rand[e.index] > p]) #scrittura molto concisa, eliminiamo i link che NON sono stati percolati

# Calculate clusters and their sizes
clusters = g.connected_components() #calcoliamo le componenti connesse (i clusters)
sizes = clusters.sizes()

sizes.sort(reverse=True)
sizes.append(0) # Questo serve unicamente a fare in modo che la size del secondo cluster, se non  proprio presente, sia 0.


print(f"First cluster size: {sizes[0]}")
print(f"Second cluster size: {sizes[1]}")
```
